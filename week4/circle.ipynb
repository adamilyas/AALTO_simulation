```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def generate_point():
    # in radians
    r = np.random.rand()
    U = np.random.rand()
    rad = 2 * np.pi * U
    x, y = np.cos(rad), np.sin(rad)
    return r*x, r*y
```

```python
points = np.transpose([generate_point() for _ in range(10000)])
plt.axis('equal')
plt.scatter(*points, s=1)
```

```python
def generate_uniform_point():
    # in radians
    r_uniform = np.sqrt(np.random.rand())
    U = np.random.rand()
    rad = 2 * np.pi * U
    x, y = np.cos(rad), np.sin(rad)
    return r_uniform*x, r_uniform*y
```

```python
points = np.transpose([generate_uniform_point() for _ in range(10000)])
plt.axis('equal')
plt.scatter(*points, s=1)
```

```python
def generate_uniform_point_r(radius=1):
    # in radians
    r_uniform = radius*np.sqrt(np.random.rand())
    U = np.random.rand()
    rad = 2 * np.pi * U
    x, y = np.cos(rad), np.sin(rad)
    return r_uniform*x, r_uniform*y
```

```python
points = np.transpose([generate_uniform_point_r(0.1) for _ in range(10000)])
plt.axis('equal')
plt.scatter(*points, s=1)
```

```python
def generate_except():
    x, y = generate_uniform_point_r(100)
    if (x-70)**2 + y**2 <= 20**2:
        x, y = generate_except()
        return x, y
    else:
        return x,y
```

```python
ls = [generate_except() for _ in range(100000)]
points = np.transpose(ls)
plt.axis('equal')
plt.scatter(*points, s=0.001)
```

```python
mn = 0
N = 100000000

for i in range(N):
    x, y = generate_except()
    mn += (transmission_rate(x,y))/N
    
    if i%(N/200)==0:
        print(i)
        
print(mn)
```

```python
def transmission_rate(x, y, r_min = 10, r_max = 100, alpha = 3, c_max = 100):
    r = np.sqrt(x**2 + y**2)
    if r <= r_min:
        return c_max
    else:
        return c_max * (r_min/ r)**alpha
```

```python
np.mean([transmission_rate(x, y) for x,y in ls])
```
